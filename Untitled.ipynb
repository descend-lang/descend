```python
%matplotlib notebook

import matplotlib.pyplot as plt
import numpy as np
```

```python
def gen_params(thread_size: int, block_size: int, grid_size: int, csv_out_path: str):
    return """
#ifndef PARAMS_CUH
#define PARAMS_CUH

#define ITEMS_PER_THREAD {0}
#define THREAD_PER_BLOCKS {1}
#define BLOCKS_PER_GRID {2}
#define CSV_OUTPUT_PATH "{3}"

#endif
    """.format(thread_size, block_size, grid_size, csv_out_path)

def write_params(params: str):
    with open("build/params.cuh", "w") as f:
        f.write(params)
```

```python
write_params(gen_params(2, 512, 32, "Benchmarking/block_size/1.csv"))
```

```python
i = 1
while i <= 1024:
    write_params(gen_params(2, i, 32, "Benchmarking/block_size/{0}.csv".format(i)))
    test = !cargo run -- run examples/infer/parallel-prefix.desc -cf cuda-examples/main.cu
    if test[-1] != "[!] Execution finished":
        print("BAD EXEC; abort")
        print(test)
        break
    else:
        print("Working {0}".format(i))
    i *= 2
```

```python
i = 1
avgs = []
while i <= 1024:
    with open("Benchmarking/block_size/{0}.csv".format(i), "r") as csvfile:
        # First 5 values tend to be longer than the other
        results = list(map(
            lambda a: float(a.split(',')[0]),
            csvfile.readlines()[1:]
        ))[0:]
        avg = sum(results) / len(results)
        avgs.append(avg)
        i *= 2

plt.plot(avgs)
plt.ylabel('Run Time of Kernel')
plt.show()
```
